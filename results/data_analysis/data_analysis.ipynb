```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import data_analysis
sns.set(font_scale=2)
```

## Analyze Induced Rotation and Translation Errors

```python
rerr, terr = data_analysis.analyze()
```

## Plot Rotation Error Distribution

```python
sns_plot = sns.kdeplot(rerr, bw_adjust=0.04, cumulative=True, common_norm=False, linewidth=3.0)
plt.xlabel('Rotation Error (rad)')
sns.move_legend(sns_plot, "lower right")
# plt.setp(sns_plot.get_legend().get_texts(), fontsize='15') 
fig = sns_plot.get_figure()
fig.savefig("rerr.png")
```

## Plot Rotation Error Distribution

```python
sns_plot = sns.kdeplot(terr, bw_adjust=0.04, cumulative=True, common_norm=False, linewidth=3.0)
plt.xlabel('Eucledian Distance (normalized to object diameter)')
sns.move_legend(sns_plot, "lower right")
# plt.setp(sns_plot.get_legend().get_texts(), fontsize='15') 
fig = sns_plot.get_figure()
fig.savefig("terr.png")
```
