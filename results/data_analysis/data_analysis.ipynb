```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import data_analysis
```

## Analyze Induced Rotation and Translation Errors

```python
rerr, terr = data_analysis.analyze()
```

## Plot Rotation Error Distribution

```python
sns.kdeplot(rerr, bw_adjust=0.04, cumulative=True, common_norm=False)
plt.xlabel('Rotation Error (rad)')
```
